# A simple imperative language

- Consider a simple language with
  - integer variable assignments
  - procedure calls with arguments
  - `if` branches (with optional `else`)
  - `while` loops
  - `print` and `input` statements

- Code blocks have mandatory brackets `{` `}`

- This language can be defined by the `Lark` grammar below

- *Note*: it is disregarding operator precedence and typing rules

```python
from lark import Lark, Transformer

grammar = """
start: block

block: stmt+

stmt: assign_stmt ";"
     | if_stmt
     | while_stmt
     | read_stmt ";"
     | print_stmt ";"

assign_stmt: CNAME "=" expr

if_stmt: "if" "(" expr ")" "{" block "}" ("else" "{" block "}")?

while_stmt: "while" "(" expr ")" "{" block "}"

print_stmt: "print" expr

read_stmt: "input" CNAME

expr: bool
     | "input"
     | CNAME -> var
     | NUMBER -> number
     | expr "+" expr -> add
     | expr "-" expr -> sub
     | expr "*" expr -> mul
     | expr "<" expr -> lt
     | expr ">" expr -> gt
     | "(" expr ")"
     | CNAME "(" args ")"

args: CNAME other_args
    |

other_args: "," CNAME other_args
           |

bool : "true" | "false"

%import common.CNAME
%import common.NUMBER
%import common.WS
%import common.BOOL
%ignore WS
"""

parser = Lark(grammar)
```
