```python
from lark import Lark, Transformer

grammar=r"""

start: func

func: "def" NAME "(" ")" "{" expr "}"

?expr: factor
     | expr SUM factor
     | expr MINUS                                                                                                                                                                     factor
     | expr MUL factor
     | expr DIV factor


?factor: NUMBER
       | paren

paren : "(" expr ")"

NUMBER: /[0-9]+/
SUM: "+"
MINUS: "-"
MUL: "*"
DIV: "/"
NAME: /[a-zA-Z]+/

%ignore /[ \t]+/
"""

parser = Lark(grammar, start="start")
```

```python
class Beautifier(Transformer):
  def start(self, items):
    return items[0]

  def func(self, items):
    return f"def {items[0]}() {{\n\n  {items[1]}\n\n}}"
  def expr(self, items):
    return f"{items[0]} {items[1]} {items[2]}"
  def NUMBER(self, items):
    return items[0]
  def factor(self, items):
    return items[0]
  def SUM(self, items):
    return items[0]
  def MINUS(self, items):
    return items[0]
  def MUL(self, items):
    return items[0]
  def DIV(self, items):
    return items[0]
  def paren(sel, items):
    return f"({items[0]})"

code = "def myFunction() { 3+5*2-(4+1) }" 
tree = parser.parse(code)

beautified = Beautifier().transform(tree)
print("Código original:", code)
print("Código formatado:\n", beautified)
```
